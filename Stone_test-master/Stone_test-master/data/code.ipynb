```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
meta = pd.read_csv("original_stone/final.csv")
print(len(meta))

meta
```

```python
13875/37 # 환자 수
```

```python
# target과 각 변수간의 상관관계

print(meta.columns)
correlation_matrix = meta.corr()['target']

# 상관 관계 출력
print(correlation_matrix)
```

```python
sorted_correlation = correlation_matrix.abs().sort_values(ascending=False)
print(sorted_correlation)
```

```python
# Assuming correlation_matrix is a Pandas Series
correlation_matrix = meta.corr()['target']

# Convert the Pandas Series to a DataFrame
correlation_matrix_df = pd.DataFrame(correlation_matrix)

# Create a heatmap from the DataFrame
plt.figure()
sns.heatmap(correlation_matrix_df, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 열 이름 수정

meta = meta.rename(columns={'Concentration (pg/um³)': 'Concentration'})
meta.columns
```

```python
# 'target'과 다른 열 간의 상관성 시각화

# 히트맵
correlation_matrix = meta.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 산점도 행렬 (선택한 열들 간의 상관 관계)
selected_columns = ['Cell Count', 'Concentration', 'Mean RI', 'Threshold RI']

sns.pairplot(meta, vars=selected_columns, hue='target', markers=['o', 's'], plot_kws={'alpha': 0.2})
plt.show()
```

```python
# 상자 그림 또는 바이올린 플롯

sns.boxplot(x='target', y='Cell Count', data= meta)
plt.title('Box Plot or Violin Plot')
plt.show()
```

```python
# 박스 플롯

ax = sns.boxplot(x='target', y='Cell Count', data=meta)
plt.title('Box Plot with Cell Count')
# sns.set(style='whitegrid')
# sns.set_palette("pastel")

# ax.legend_.remove()
plt.show()
```

```python
# 열 선택
selected_columns = ['Cell Count', 'Concentration', 'Mean RI', 'Threshold RI']

# 박스 플롯 그리기
plt.figure(figsize=(14, 6))  # 그래프 전체 크기 조절
sns.boxplot(x='target', y='value', hue='variable', data=pd.melt(meta, id_vars=['target'], value_vars=selected_columns), palette='pastel')
plt.title('Box Plots by Target')
plt.show()
```

```python
columns = ['Concentration', 'Mean RI', 'Threshold RI']


plt.figure(figsize=(14, 6))  # 그래프 전체 크기 조절
sns.boxplot(x='target', y='value', hue='variable', data=pd.melt(meta, id_vars=['target'], value_vars= columns), palette='pastel')

# y축 범위 제한
ax.set_ylim(0, 150)

plt.title('Box Plots by Target')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 예시 데이터프레임
# 'target', 'Cell Count', 'Concentration', 'Mean RI', 'Threshold RI' 등 열이 포함된 meta 데이터프레임을 가정합니다.

# 열 선택
selected_columns = ['Cell Count', 'Concentration', 'Mean RI', 'Threshold RI']

# 그래프 전체 크기 조절
plt.figure(figsize=(14, 10))

# 각 열에 대한 개별적인 box plot 그리기
for i, col in enumerate(selected_columns, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='target', y=col, data=meta, width=0.4, palette='pastel')
    plt.title(f'Box Plot of {col} by Target')

plt.tight_layout()
plt.show()
```
